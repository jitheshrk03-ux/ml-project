```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
seoul_bike_sharing_demand = fetch_ucirepo(id=560)

# data (as pandas dataframes)
x = seoul_bike_sharing_demand.data.features
y = seoul_bike_sharing_demand.data.targets
```

```python
import pandas as pd

df=pd.concat([x,y],axis=1)
```

```python
df.columns=df.columns.str.lower()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df[['day','month','year']]=df['date'].str.split('/',expand=True)
```

```python
df[['day','month','year']]=df[['day','month','year']].astype(int)
```

```python
df.drop('date',axis=1,inplace=True)
```

```python

from sklearn.model_selection import train_test_split
```

```python
non_functional_index = df['functioning day'] == 'Yes'
cleared_data = df.loc[non_functional_index]
```

```python
target_column='rented bike count'# name of the columns to predict
features=cleared_data.drop(columns=[target_column]) # every other column expect target
target=cleared_data[target_column]
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2)
```

```python
len(x_train) == len(y_train)
```

```python
len(x_test) == len(y_test)
```

```python
from sklearn.compose import make_column_selector
from sklearn.compose import ColumnTransformer
```

```python
numeric_selector=make_column_selector(dtype_include='number') # only selects numeric col in features
categorical_selector=make_column_selector(dtype_exclude='number')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
categorical_columns=categorical_selector(features)
categorical_columns
```

```python
numeric_columns=numeric_selector(features) #extract names of the numeric column
numeric_columns
```

```python
preprocessor=ColumnTransformer([
    ('category',OrdinalEncoder(),categorical_columns),
    ('numbers',StandardScaler(),numeric_columns)
])
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
linear_regresion=Pipeline([
    ('preprocessor',preprocessor),
    ('gbr',GradientBoostingRegressor(
        n_estimators=100,
        learning_rate=0.1,max_depth=3,
        random_state=42
    ))
])
```

```python
linear_regresion.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred_train=linear_regresion.predict(x_train)
```

```python
import numpy as np
mae_train=mean_absolute_error(y_train,y_pred_train)
mse_train=mean_squared_error(y_train,y_pred_train)
rmse_train=np.sqrt(mse_train)
r2_score_train=r2_score(y_train,y_pred_train)
```
